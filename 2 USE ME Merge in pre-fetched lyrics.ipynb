```python
import pandas as pd
import numpy as np
import json

from IPython.display import Markdown as md
```

# 1. load in data

```python
songs_df = pd.read_csv('./data/1 DONE RIGHT OUTPUT unique songs.csv')
songs_df
```

```python
LYRIC_FOLDER_PATH = './data/lyric-fetching-datasets'
```

```python
lyrics_df = pd.read_csv(f'{LYRIC_FOLDER_PATH}/{"ALL LYRICS DATASET (FINAL.. for now *cries*)"}.csv')
lyrics_df
```

# 2. Merge (note we'll drop ~60 tracks identified as instrumental whe creating the lyrics dataset)

```python
merged = songs_df.merge(
    right=lyrics_df,
    on='song_id',
    how='inner'
)

merged
```

# 3. Export!

```python
merged.to_csv('./data/2 OUTPUT Songs with lyrics.csv', index=False)
```
