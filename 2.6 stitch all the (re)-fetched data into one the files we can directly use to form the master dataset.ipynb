# 2️⃣ SECOND: stitch

How i propose:
0. start with the base dataset
1. UPDATE PHASE: we want to append or *replace* newly fetched data from the 2 datasets
2. FILTER OUT INSTRUMENTALS
3. check our coverage, export as lean as possible for merging

```python
JUST_SONG_ID_AND_LYRICS_COLS = ['song_id', 'lyrics']
SONG_ID = 'song_id'

def remove_overlap(base_df, overlap_df_to_remove):
    return base_df[
        base_df.song_id.apply(
            lambda i: i not in overlap_df_to_remove.song_id.values
        )
    ]
```

## A. Ok, the 69 improved need to be swapped in:

```python
better_url_plus_base_df = pd.concat([
    remove_overlap(
        base_df,
        better_url_refetched_lyrics
    ),
    better_url_refetched_lyrics[JUST_SONG_ID_AND_LYRICS_COLS]
])

better_url_plus_base_df
```

## B. Now add refetched ~420

```python
big_refetch_of_misc
```

```python
big_refetch_of_misc[big_refetch_of_misc.lyrics.isna()]
```

```python
# to test overlap:
# better_url_plus_base_df.merge(on='song_id', right=misc_non_null) 

misc_plus_better_url_plus_base_df = pd.concat([
    remove_overlap(
        better_url_plus_base_df,
        big_refetch_of_misc
    ),
    big_refetch_of_misc
])

misc_plus_better_url_plus_base_df.drop_duplicates(subset=['song_id']).dropna(subset=['song_id', 'lyrics']) # just to prove we dont' need it :)
```

```python
final_with_instrumentals_removed = remove_overlap(
    misc_plus_better_url_plus_base_df,
    instrumentals_so_no_lyrics
)
final_with_instrumentals_removed
```

# 3️⃣ Now just save the lyrics by `song_id`!

```python
def save_lyrics(lyrics_subset, file_name):
    lyrics_subset.to_csv(f'{LYRIC_FOLDER_PATH}/{file_name}.csv', index=False)
    return lyrics_subset
```
